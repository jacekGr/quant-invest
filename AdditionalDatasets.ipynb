## Exchange Rates
Daily exchange rates relative to SDR from IMF (https://www.imf.org/external/np/fin/data/param_rms_mth.aspx)
May contain NAs

```python
exchange_rates = pd.read_csv('input/exchange_rates.csv', index_col = 0)
exchange_rates.index = pd.to_datetime(exchange_rates['Date'])
exchange_rates['Date']= pd.to_datetime(exchange_rates['Date'])
shortnames={}
for col in exchange_rates.columns[1:]:
    start = col.index('(')
    end = col.index(')')
    shortnames[col] = col[(start+1):end]
exchange_rates = exchange_rates.rename(columns = shortnames)
```

example:

```python
exchange_rates.iloc[-5:]['USD']/exchange_rates.iloc[-5:]['EUR']
```

## Interest Rates
Daily Interest Rates from https://www.bis.org, using https://cran.r-project.org/web/packages/BIS/index.html. 

Description: https://www.bis.org/statistics/bulletin1812.pdf#page=340

```python
policy_rates = pd.read_csv('input/policy_rates.csv', index_col = 0)
policy_rates = policy_rates.pivot(index = 'date', columns = 'reference_area', values = 'obs_value')
policy_rates = policy_rates.iloc[:(-42)]
```
